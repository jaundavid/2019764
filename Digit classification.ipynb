```python
import numpy as np
import pandas as pd

train = pd.read_csv('Descargas/train.csv')
test = pd.read_csv('Descargas/test.csv')
```

```python
X = np.concatenate((train5.values[:2656], train8.values[:2844]))
```

```python
a=np.zeros(2656)
b=np.ones(2844)
y = np.concatenate((a,b))
y
```

```python
from sklearn.svm import SVC
clf = SVC()
clf.fit(X, y) 
```

```python
X[2]
```

```python
print(clf.predict([X[2]]))
```

```python
from sklearn import cross_validation
from sklearn import svm
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.4, random_state=23)
```

```python
clf = svm.SVC(kernel='linear', C=1).fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
def exploration_parameters(X_train, y_train, X_test, y_test):
    pscores = []
    Cparameters = []
    for a in range(-5,16):
        Cparameters.append(2 ** a)
    for c in range(len(Cparameters)):
        clf = svm.SVC(kernel='linear', C=Cparameters[c]).fit(X_train, y_train)
        pscores.append(clf.score(X_test, y_test))
    return pscores, Cparameters
```

```python
scores, c_parameters = exploration_parameters(X_train, y_train, X_test, y_test)
```

```python
scores, c_parameters
```

```python
testscores = []
Cparameters = []
a=np.zeros(1139)
b=np.ones(1219)
ytest = np.concatenate((a, b))
for a in range(-5,16):
    Cparameters.append(2 ** a)
for c in range(len(Cparameters)):
    clf = svm.SVC(kernel='linear', C=Cparameters[c]).fit(X_train, y_train)
    testscores.append(clf.score(np.concatenate((train5.values[2656:], train8.values[2844:])), ytest))
```

```python
testscores, Cparameters
```

Below we plot the training error vs test error 

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(c_parameters, scores, '^r', label="Training")
plt.plot(Cparameters, testscores, '*', label="Test")
plt.legend()
plt.xscale('log', basex=2)
plt.xlabel("C")
plt.ylabel("Accuracy")
```

As we can see above the C complexity parameter did not play a relevant role during the training and the validation stage. It could be due to the separation between number five and eight is very clear and the importance given to the complexity parameter is less significant in this case.
