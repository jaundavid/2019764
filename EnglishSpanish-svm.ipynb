```python
def edit(px, py):
    
    len1 = len(px)
    len2 = len(py)
    row = np.zeros((2,len1+1))
    
    for a in range(len1+1):
        row[0][a] = a
    now = 1
    old = 0
    
    for a in range(len2):
        row[now][0] = a+1
        for b in range(len1):
            row[now][b+1] = min(min(row[old][b] + (0 if px[b]==py[a] else 1), row[old][b+1] + 1), row[now][b] + 1)
        now = 1-now
        old = 1-old
    result=row[old][len1];
    return result
edit("archacho","rcpecho")
```

(b) Program a string kernel.

In order to program the new kernel we have inherited the svm.SVC class and modified the fit() and predict() methods. Our Kernel function is defined as follows:

$K : \sum^d \times \sum^d \Rightarrow \mathbb{R}$

where 

$x, y \Rightarrow e^{-gamma \times edit(x, y)^2}$

We have used as gamma parameter $1$

```python
LIBSVM_IMPL = ['c_svc', 'nu_svc', 'one_class', 'epsilon_svr', 'nu_svr']

class StringKernelSVM(svm.SVC):
    def __init__(self, complexity):
        """
        Constructor
        :param lambda_decay: lambda parameter for the algorithm
        :type  lambda_decay: float
        :param subseq_length: maximal subsequence length
        :type subseq_length: int
        """
        svm.SVC.__init__(self, kernel='precomputed', C=complexity)
    
    def string_kernel(self, K, G):
        R = np.ones((len(K), len(G)))
        if len(K) != len(G):
            for a in range(len(K)):
                for b in range(len(G)):
                    R[a][b] = scipy.exp(-int(levenshtein(K[a] , G[b]) ** 2))
        else:
            for a in range(len(K)):
                for b in range(a, len(G)):
                    R[a][b] = scipy.exp(-int(levenshtein(K[a] , G[b]) ** 2))
                    R[b][a] = R[a][b]
        
        return R
    
    def fit(self, X, Y):
        gram_matr = self.string_kernel(X, X)
        self.__X = X
        super(svm.SVC, self).fit(gram_matr, Y)
        
    def predict(self, X):
        svm_type = LIBSVM_IMPL.index('c_svc')
        if not self.__X:
            print('You should train the model first!!!')
            sys.exit(3)
        else:
            gram_matr_predict_new = self.string_kernel(X, self.__X)
            gram_matr_predict_new = np.asarray(gram_matr_predict_new, dtype=np.float64, order='C')
            return libsvm.predict(
                gram_matr_predict_new, self.support_, self.support_vectors_, self.n_support_,
                self.dual_coef_, self._intercept_,
                self._label, self.probA_, self.probB_,
                svm_type=svm_type,
                kernel=self.kernel, degree=self.degree, gamma=self._gamma, coef0=self.coef0,
                cache_size=self.cache_size)
    
        
```

(d) Use cross validation to find an appropiate complexity parameter.

In order to do cross validation, we have used too scikit-learn cross validation module. Concretely we have used the 15000 most frequent words(7500 of english and 7500 of spanish), then we splitted the first 2500 words of each language as a training set and the other 10000 words were used as test data. Using KFold we got the indices for the 5-fold cross validation.

Kfold returns two arrays, the first one containing the training indices and the second one containing the validation indices.

```python
from sklearn.cross_validation import KFold
kf = KFold(5000, n_folds=5, shuffle=True, random_state=45)
for train, test in kf:
    print("%s %s" % (train, test))
X = ing4.word[:2500].tolist() + esp4.word[:2500].tolist()
X = np.array(X)
a = np.zeros(2500)
b = np.ones(2500)
y = np.concatenate((a, b))
for train, test in kf:
    print X[train]
    print X[test]
```

In order to get the accuracy we define the following function that takes the number of correct predictions over the total number of predictions. Then we iterate over the 5 folds and append the results to a scores list that keeps the accuracy results.

```python
def get_accuracy(result, ytest):
    correct = 0.0
    for a in range(len(result)):
        if result[a] == ytest[a]:
            correct += 1
    return correct / len(result)
scores = []
for train, test in kf:
    clf = StringKernelSVM(1)
    clf.fit(X[train].tolist(), y[train])
    result = clf.predict(X[test].tolist())
    scores.append(get_accuracy(result, y[test]))
```

This is an example using $C=1$ as parameter for the SVC classifier.

```python
scores
```

Finally we define a function that iterates over $\{2^{-5},2^{-4},2^{-3},2^{-2},2^{-1},2^0,2^1,2^2,2^3,2^4,2^5 \}$ as the complexity parameter and returns the scores for each C and for each fold respectively

```python
def exploration_parameters(kf, X, y):
    pscores = []
    Cparameters = []
    for a in range(-4,6):
        pscores.append([])
        Cparameters.append(2 ** a)
    for c in range(len(Cparameters)):
        for train, test in kf:
            clf = StringKernelSVM(Cparameters[c])
            clf.fit(X[train].tolist(), y[train])
            result = clf.predict(X[test].tolist())
            pscores[c].append(get_accuracy(result, y[test]))
    return pscores, Cparameters
```

```python
scores, c_paramaters = exploration_parameters(kf, X, y)
```

```python
scores,c_paramaters
```

In the figure below we find that $C=1$ is the optimal parameter.

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(c_paramaters, scores, '*')
plt.xscale('log', basex=2)
plt.xlabel("C")
plt.ylabel("Accuracy")
```
