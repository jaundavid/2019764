For this purpose we used the following library for Python: https://github.com/eBay/bayesian-belief-networks. It allows us to model a bayesian belief network. Firstly we import the library

```python
from bayesian.bbn import build_bbn
```

To model a bayesian belief network we need to declare each random variable as a function. In our case the model is compound of five variables: The season, the locale, the type of fish, the lightness of the fish and the thickness of the fish. Below we define for each variable a function, that represents the value the variable could take. For instance in the case of the seasons, each season has got the same probability so it should return $0.25$

```python
def f_seasons(seasons):
    return 0.25
```

Then we define the function for locale. The definition is slightly different to the previous one. Here we have to compare whether the locale is in the north atlantic sea or south atlantic. Later when we define the context of each variable, we set the value that each variable could take.

```python
def f_locale(locale):
    if locale == "north":
        return 0.6
    else:
        return 0.4
```

The variable fish is particular because the season and the locale determine directly the probability of two different types of fish.(Salmon and sea bass) And we encode the information of this conditional probability as a function. We do the same process for lightness and thickness.

```python
def f_fish(seasons,locale,fish):
    if seasons == "winter" and locale == "north":
        return 0.5
    elif seasons == "winter" and locale == "south":
        if fish == "salmon":
            return 0.7
        else:
            return 0.3
    elif seasons == "spring" and locale == "north":
        if fish == "salmon":
            return 0.6
        else:
            return 0.4
    elif seasons == "spring" and locale == "south":
        if fish == "salmon":
            return 0.8
        else:
            return 0.2
    elif seasons == "summer" and locale == "north":
        if fish == "salmon":
            return 0.4
        else:
            return 0.6
    elif seasons == "summer" and locale == "south":
        if fish == "salmon":
            return 0.1
        else:
            return 0.9
    elif seasons == "autumn" and locale == "north":
        if fish == "salmon":
            return 0.2
        else:
            return 0.8
    elif seasons == "autumn" and locale == "south":
        if fish == "salmon":
            return 0.3
        else:
            return 0.7

def f_lightness(fish,lightness):
    if fish == "salmon":
        if lightness == "light":
            return 0.6
        elif lightness == "medium":
            return 0.2
        else:
            return 0.2
    else:
        if lightness == "light":
            return 0.2
        elif lightness == "medium":
            return 0.3
        else:
            return 0.5

def f_thickness(fish,thickness):
    if fish == "salmon":
        if thickness == "wide":
            return 0.3
        else:
            return 0.7
    else:
        if thickness == "wide":
            return 0.6
        else:
            return 0.4
```

## Statement of the model

In order to define the model, we call the function *build_bbn* and we pass it the functions we recently defined as parameters. Additionally we state the domain of each variable.

```python
if __name__ == '__main__':

    g = build_bbn(
        f_seasons,
        f_locale,
        f_fish,
        f_lightness,
        f_thickness,
        domains=dict(
            seasons=['winter', 'spring', 'summer', 'autumn'],
            locale=['north', 'south'],
            fish=['salmon', 'sea bass'],
            lightness=['light', 'medium', 'dark'],
            thickness=['wide', 'thin']))
```

It calculates automatically the marginal probability of each random variable. After we state the model, we can query it as follows:

```python
g.q()
```

a.) A dark, thin fish is caught in the north Atlantic in summer. What is the probability it is a salmon?  
We can formulate it as a query where we define the values of some variables:

```python
g.q(lightness='dark', locale='north', thickness='thin')
```

This probability is equal to $0.340974$

b.) A thin, medium fish is caught in the north Atlantic. What is the probability it is winter? spring? summer? autumn?

```python
g.q(lightness='medium', locale='north', thickness='thin')
```

This is called casual inference and as we can see the probabilities are slightly different given the fact that its lightness and thickness are already known.

c.) A light, wide fish is caught in the autumn. What is the probability it came from the north Atlantic?

```python
g.q(lightness='light', seasons='autumn', thickness='wide')
```

In this case the probability of north decreases a little bit from $0.60$ to $0.589286$
