#Dataset
The structure of dataset ([data.txt](https://sites.google.com/a/unal.edu.co/machine-learning-2013-2/data.txt?attredirects=0)) is $ x_i y_i \mathcal{C_i} $ where $(x_i,y_i)\in\mathbb{R}^2$ and $\mathcal{C_i} \in \{0,1,2\}$

```python
import numpy as np
import pylab as pl

data = np.loadtxt("data.txt", delimiter=",")

D = data[:,2]
x = data[:,0:2]

ind = [[]]*3
c = [[]]*3

ind[0] = (D==0).nonzero()
ind[1] = (D==1).nonzero()
ind[2] = (D==2).nonzero()
c[0] = x[ind[0]]
c[1] = x[ind[1]]
c[2] = x[ind[2]]

pl.scatter(c[0][:,0],c[0][:,1],color='cyan',marker='o',label='$C_0$')
pl.scatter(c[1][:,0],c[1][:,1],color='magenta',marker='^',label='$C_1$')
pl.scatter(c[2][:,0],c[2][:,1],color='yellow',marker='s',label='$C_2$')
pl.ylabel('$x_2$', size=20)
pl.xlabel('$x_1$', size=20)
pl.legend()
```


# Parameter estimation
Given that:

* Data are sampled form a probability distribution $p_k(x,y)$ for each class $(k \in \{0,1,2\})$.
* The form of the probability distribution $p$ is know for each class $(\mathbf{x} \sim \mathcal{N}_2(\boldsymbol\mu, \boldsymbol\Sigma))$ but its parameters are unknown.
* There is a training set $\mathcal{C}_k = \{(x_1,y_1)_k,\ldots,(x_n,y_n)_k\}$ of examples sampled *i.i.d* according to $p_k(x,y)$.

We must estimate the unkown parametrs of each $p_k$, i.e. $\boldsymbol\mu$ (the mean or location vector $\in \mathbb{R}^2$ and $\boldsymbol\Sigma$ covariance nonnegative-definite matrix $\in \mathbb{R}^{2\times 2}$) from training data $\mathcal{C}$.

The parameters can be written as: $
\boldsymbol\mu =
\begin{pmatrix}
\mu_x \\
\mu_y 
\end{pmatrix}, \quad
\boldsymbol\Sigma =
\begin{pmatrix}
\sigma_x^2 & \rho \sigma_x \sigma_y \\
\rho \sigma_x \sigma_y  & \sigma_y^2 \\
\end{pmatrix}.
$

In order to estimate the parameters of each class we use $80\%$ of the examples choosen randomly:

```python
n = len(c[0])
train = [[]]*3
test = [[]]*3
percent = int(n*0.8)
for i in xrange(0,3):
    rand_indices = np.random.permutation(n)
    train_indices = rand_indices[:percent]
    test_indices = rand_indices[percent:]
    train[i] = c[i][train_indices]
    test[i] = c[i][test_indices]
```

# Maximum likelihood estimation
- Assumes parameters fixed but unknown values
- Values are computed as those maximizing the probability of observerd examples $\mathcal{C_i}$ (the training set for the class).
- Obtained values are used to compute probability for new examples.

The maximum likelihood covariance estimator in scikitlearn is known as empirical covariance and can be computed by fitting an EmpiricalCovariance object to the data with the fit method.

```python
from sklearn.covariance import EmpiricalCovariance
model = [[]]*3
mu = [[]]*3
for i in xrange(0,3):
    model[i] = EmpiricalCovariance()
    model[i].fit(train[i])
    print "Sigma_",i,"= "
    print model[i].covariance_
```

The MLE for $\boldsymbol\mu$ for the case of bivariate gaussian distribution is given for $\hat{\boldsymbol\mu}=\begin{pmatrix} \frac{1}{n} \sum_i x_1^{(i)} \\ \frac{1}{n} \sum_i x_2^{(i)} \end{pmatrix}$

```python
for i in xrange(0,3):
    mu[i] = np.average(train[i],0)
    print "mu_",i,"= "
    print mu[i]
```

# Multivariate classification

$$ p(\mathcal{C}_i|x) = \frac{P(\mathcal{C}_i)p(x|\mathcal{C}_i)}{p(x)} = \frac{P(\mathcal{C}_i)p(x|\mathcal{C}_i)}{\sum_k p(x|\mathcal{C}_k)P(\mathcal{C}_k)} $$

- $P(\mathcal{C}_i)$ is called the **prior probability** that $\mathcal{C}$ takes the value of $i$, regardless of the $x$ value, It is called the prior because it is the knowledge we have as to the value of $\mathcal{C}$ *before* looking at observables $x$, satisfying $ P(\mathcal{C}_i) \geq 0 \text{ and } \sum_k{P(\mathcal{C}_i)}=1 $

- $p(x|\mathcal{C}_i)$ is called the **class likelihood** and is the conditional probability than an event belonging to $\mathcal{C}_i$ has the associated observation value $x$. if te class-conditional densities are taken as normal densities, we have: $ p(x|\mathcal{C}_i)=\frac{1}{(2\pi|\boldsymbol\Sigma_i|)^{1/2}}exp[-\frac{1}{2}(x-\boldsymbol\mu_i)^T \boldsymbol\Sigma_i^{-1}(x-\boldsymbol\mu_i)]$

- $p(x)$, the **evidence**, is the marginal porbability that an observation $x$ is seen, regardless of the class.

Once we have the posteriors, we decide the class with the highest posterior probability; that is, we choose $\mathcal{C}_i$ if $P(\mathcal{C}_i|x) = \max_k P(\mathcal{C}_k|x)$.

```python
from numpy.linalg import det
from mpl_toolkits.mplot3d import axes3d, Axes3D

#In our case this function is not necessary because priors and evidence has uniform probabilities within classes.
def posteriori(prior,likelihood,evidence):
    return (prior*likelihood)/evidence

def likelihood(model,mean,x):
    return np.exp(-0.5*np.dot(np.dot((x-mean),model.precision_),(x-mean)))/np.sqrt(2*np.pi*det(model.covariance_))
```

We define the discriminant function as $g_i(x) = \log(p(x|\mathcal{C}_i))+log(P(\mathcal{C}_i))$ and since $p(x|\mathcal{C_i}) \sim \mathcal{N}_2(\boldsymbol\mu_i,\boldsymbol\Sigma_i)$, we have

$$
g_i(x) = -\frac{1}{2}\log(2\pi)-\frac{1}{2}\log(|\boldsymbol\Sigma_i|)-\frac{1}{2}(x-\boldsymbol\mu_i)^T\boldsymbol\Sigma_i^{-1}(x-\mu_i)+ \log(P(\mathcal{C}_i))
$$
And the estimators are plugged into the discriminant function:
$$
g_i(x) = -\frac{1}{2}\log(2\pi)-\frac{1}{2}\log(|\hat{\boldsymbol\Sigma}_i|)-\frac{1}{2}(x-\hat{\boldsymbol\mu}_i)^T\hat{\boldsymbol\Sigma}_i^{-1}(x-\hat{\boldsymbol\mu}_i)+ \log(\hat{P}(\mathcal{C}_i))
$$

Expanding
$$
g_i(x)=-\frac{1}{2}\log(|\hat{\boldsymbol\Sigma}_i|)-\frac{1}{2}(x^{T}\hat{\boldsymbol\Sigma}_i^{-1}x-2x^T\hat{\boldsymbol\Sigma}_i^{-1}\hat{\boldsymbol\mu}_i+\hat{\boldsymbol\mu}_i^T\hat{\boldsymbol\Sigma}_i^{-1}\hat{\boldsymbol\mu}_i)+ \log(\hat{P}(\mathcal{C}_i))
$$

which defines a *quadratic discriminant* that can also be written as: $g_i(x)=x^T \boldsymbol{W}_i x + \boldsymbol{w}_i^Tx + w_i$ where:
$$
\boldsymbol{W}_i = -\frac{1}{2}\hat{\boldsymbol\Sigma}_i^{-1} \\
\boldsymbol{w}_i = \hat{\boldsymbol\Sigma}_i^{-1}\hat{\boldsymbol\mu}_i \\
w_i = - \frac{1}{2}\hat{\boldsymbol\mu}_i^{T} \hat{\boldsymbol\Sigma}_i^{-1}\hat{\boldsymbol\mu}_i - \frac{1}{2}\log(|\hat{\boldsymbol\Sigma}_i|)+ \log(\hat{P}(\mathcal{C}_i))
$$
We can drop the prior probabilities (since we have equals priors in this case), and ignore the $-\frac{1}{2}\log(2\pi)$ constant term:

```python
def discriminant_function(model,mean,x):
    return -0.5*log(det(model.covariance_))-0.5*np.dot(np.dot((x-mean),model.precision_),(x-mean))
```
